```python
fp = open("sample1.txt","w")
fp.write("Hello python")
content =  fp.read()
print(content)
fp.close()
```

```python
fp = open("sample1.txt","w+")
print(fp.tell())
fp.write("Hello python")
print(fp.tell())
fp.seek(0,0)
print(fp.tell())

content =  fp.read()
print(fp.tell())
print(content)
fp.close()
```

```python
tell() : current cur position
    
seek(offset,position)

0 : start of the file 
1 : current position 
2 : end of the file 
    
offset : how many chars
    
    
seek(15,0) : 15 char from start of the file 
seek(0,0) : 0 char from start of the file => Start of the file 
seek(0,2) : 0 chars from end of the file => end of the file  
    
    
seek(15,1) : error
seek(15,2) : error

    
    seek(0,1)
    seek(0,2)
```

```python
fp = open("sample1.txt","r+")
fp.seek(0,2)
fp.write("\nHelllooooooooo")
fp.close()
```

```python
fp = open("sample.txt","a+",encoding="utf-8")
print(fp.tell())
fp.write("\n\n Add to my file ")
fp.seek(0,0)
content = fp.read()
print(content)
fp.close()
```

```python
r : fp => start of the file, read opeartion, file does not exist : error
r+ : fp => start , read + write , file does not exist : error

w : fp => start , write, create new 
w+ : fp => start, write + read , create new file 

a : fp => end of the file, write ,create new file 
a+: fp > end, opeartions : write + read , craete new file 
```

```python
seek() : change the fp position
tell : get the current fp position.
```
