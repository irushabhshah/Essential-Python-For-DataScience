```python
def fibo():
    
    num1 = 0
    yield num1
    num2 = 1
    yield num2
    while(True):
        next_num = num1 + num2
        yield next_num
        num1,num2 = num2,next_num
        
g = fibo()
print(g)
print(next(g))
print(next(g))
print(next(g))
print(next(g))
# print(next(g))
```

```python
def fibo():
    nums = [0,1]
    count = 2
    num1 = 0
    num2 = 1
    while(True):
        next_num = num1 + num2
        nums.append(next_num)
        count+=1
        if count == 5:
            yield nums
            count = 0
            nums.clear()
        num1,num2 = num2,next_num
        
g = fibo()
print(g)
print(next(g))
print(next(g))
print(next(g))
print(next(g))
# print(next(g))
```

```python
def printval(l):
    for value in l:
        if value == 30:
            return value
        yield value
    
l = [10,20,30,40,50]
g = printval(l)
print(next(g))
print(next(g))
print(next(g))
# print(next(g))
# print(next(g))
# print(next(g))
```

```python
def add_deco(func):
    def new_func(num1,num2):
        if type(num1) == type(num2):
            return func(num1,num2)
        else:
            return func(str(num1),str(num2))
        
    return new_func 
    
    
    
@add_deco
def add(num1,num2):
    return num1 + num2

print(add(10,20))
print(add("Python","Java"))
print(add("Python",100))
```

```python
add(10,20) : - @add_deco 
             - def add_dec(func = add):
             - return new_func => return 30 
             - def new_func(10,20):
                    return func(10,20) => add(10,20) => return 30 
             - def add(num1,num2):
                   return 30 
            
```

```python
def check_login(func):
    def login(username,password,val):
        if username== "abc" and password == "abc@123":
            return func(username,password,val)
        else:
            return False    
    return login

@check_login
def edit_name(username,password,new_name):
    print("Inside edit name")
    return new_name
    
    
@check_login    
def edit_phone_number(username,password,new_number):
    print("Inside edit phone number")
    return new_number
    
@check_login
def edit_email(username,password,new_email):
    print("Inside edit email")
    return new_email


print(edit_name("abc","abc@12","pqr"))
# print(edit_email("abc","abc@124","xyz@pqr.com"))
```

```python
def check_login(func):
    def login(*args,**kwargs):
        username = kwargs.get('username')
        password = kwargs.get('password')
        new_name = kwargs.get('new_name')
        new_email = kwargs.get('new_email')
        
        if kwargs.get('username')== "abc" and kwargs.get('password') == "abc@123":
            if new_name:
                return func(username,password,new_name)
            else:
                return func(username,password,new_email)
        else:
            return False    
    return login

@check_login
def edit_name(username,password,new_name):
    print("Inside edit name")
    return new_name
    
    
@check_login    
def edit_phone_number(username,password,new_number):
    print("Inside edit phone number")
    return new_number
    
@check_login
def edit_email(username,password,new_email):
    print("Inside edit email")
    return new_email


print(edit_name(username= "abc",password= "abc@12",new_name = "pqr"))
# print(edit_email("abc","abc@124","xyz@pqr.com"))
```
